```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.restoration import inpaint
from skimage.transform import resize
from skimage import color
from skimage import io, filters, measure, morphology,restoration
from skimage.util import random_noise
from PIL import Image
from skimage.util import random_noise
from skimage.restoration import denoise_tv_chambolle
from skimage.restoration import denoise_bilateral
from skimage.segmentation import slic
from skimage.color import label2rgb
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('on')
    
def plot_comparison(img_original, img_filtered, img_title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 10), sharex=True, sharey=True)
    ax1.imshow(img_original, cmap=plt.cm.gray)
    ax1.set_title('Original')
    ax1.axis('on')
    ax2.imshow(img_filtered, cmap=plt.cm.gray)
    ax2.set_title(img_title_filtered)
    ax2.axis('on')
def calculate_pixels(image_path):
    image = Image.open(image_path)
    image_array = np.array(image)
    height, width, channels = image_array.shape
    pixels = height * width
    return pixels    
```

```python
#Câu 1: Khôi phục ảnh bị hỏng

image_path = "anhtest.jpg" 
image = Image.open(image_path)


image_array = np.array(image) 
image_array[100:300, 200:400] = [0, 0, 0] 

mask = np.zeros(image_array.shape[:2], dtype=bool)
mask[100:300, 200:400] = True  

restored_image = restoration.inpaint.inpaint_biharmonic(image_array, mask, channel_axis=-1)

plot_comparison(image_array,restored_image , 'Restored image')
```

```python
#Câu 2: Xóa một logo không mong muốn ra khỏi ảnh ban đầu.

image_with_logo = plt.imread('anhtestc3.jpg')
mask = np.zeros(image_with_logo.shape[:-1])

mask[1050:1580, 250:1550] = 1
mask[2000:2200, 250:1000] = 1
mask[1700:2200, 900:1550] = 1
mask[1700:2200, 1583:] = 1

image_logo_removed = inpaint.inpaint_biharmonic(image_with_logo,
                                                mask,
                                                channel_axis=-1)

plot_comparison(image_with_logo, image_logo_removed, 'Image with dots removed')
```

```python
#Câu 3: Khởi tạo ảnh nhiễu


image = plt.imread('anhtest.jpg')

noisy_image1 = random_noise(image,mode = 'speckle')
noisy_image2 = random_noise(image,mode = 'localvar')
noisy_image3 = random_noise(image,mode = 'poisson')
noisy_image4 = random_noise(image,mode = 'salt')
plot_comparison(image, noisy_image1, 'Speckle')
plot_comparison(image, noisy_image2, 'Localvar')
plot_comparison(image, noisy_image3, 'Poisson')
plot_comparison(image, noisy_image4, 'Salt')

h1,h2,c1,c2 = 100,200, 200, 400 
plot_comparison(image[h1:h2,c1:c2, : ], noisy_image1[h1:h2,c1:c2, : ], 'Speckle')
plot_comparison(image[h1:h2,c1:c2, : ], noisy_image2[h1:h2,c1:c2, : ], 'Localvar')
plot_comparison(image[h1:h2,c1:c2, : ], noisy_image3[h1:h2,c1:c2, : ], 'Poisson')
plot_comparison(image[h1:h2,c1:c2, : ], noisy_image4[h1:h2,c1:c2, : ], 'Salt')
```

```python
#Câu 4: Giảm nhiễu bằng phương pháp Chambolle

image = plt.imread('anhtest.jpg')
noisy_image = random_noise(image,mode = 'salt')
denoised_image = denoise_tv_chambolle(noisy_image, channel_axis=2)
plot_comparison(noisy_image, denoised_image, 'Denoised Image with chambolle')
h1,h2,c1,c2 = 100,200, 200, 400 
plot_comparison(noisy_image[h1:h2,c1:c2, : ], denoised_image[h1:h2,c1:c2, : ], 'Denoised Image with chambolle')
```

```python
#Câu 5:  Giảm nhiễu bằng bộ lọc bilataral filter

image = plt.imread('anhtest.jpg')
noisy_image = random_noise(image,mode = 'salt')
denoised_image = denoise_bilateral(image, channel_axis=2)

plot_comparison(noisy_image, denoised_image, 'Denoised Image with bilateral')

h1,h2,c1,c2 = 100,200, 200, 400 
plot_comparison(noisy_image[h1:h2,c1:c2, : ], denoised_image[h1:h2,c1:c2, : ], 'Denoised Image with bilateral')
```

```python
#Câu 6: Tính tổng số pixels của ảnh

image_path = "anhtest.jpg"
num_pixels = calculate_pixels(image_path)
print("Number of pixels in the image:", num_pixels)
image = np.array(Image.open(image_path))
show_image(image)
```

```python
#Câu 7: Giảm số điểm ảnh của ảnh xuống ngưỡng 400 siêu điểm ảnh

face_image = plt.imread('mat2.jpg')

segments = slic(face_image, n_segments=400)

segmented_image = label2rgb(segments, face_image, kind='avg')

plot_comparison(face_image, segmented_image, 'Segmented image, 400 superpixels')
```

```python
#Câu:  Hình dạng đường bao của ảnh nhị phân
image_path = "anhtest_binary.jpg"
image = io.imread(image_path, as_gray=True)

threshold_value = filters.threshold_otsu(image)
binary_image = image > threshold_value

contours = measure.find_contours(binary_image, 0.2)

min_area = 200

fig, ax = plt.subplots()
ax.imshow(image, cmap=plt.cm.gray)

for contour in contours:
    contour_area = len(contour)
    if contour_area > min_area:
        ax.plot(contour[:, 1], contour[:, 0], linewidth=2, color='green')
```

```python
#Câu 9: Tìm đường bao của một ảnh màu

image_path = "anhtest.jpg"
image = io.imread(image_path, as_gray=True)

threshold_value = filters.threshold_otsu(image)
binary_image = image > threshold_value

contours = measure.find_contours(binary_image, 0.8)

min_area = 800  

fig, ax = plt.subplots()
ax.imshow(image, cmap=plt.cm.gray)

for contour in contours:
    contour_area = len(contour)
    if contour_area > min_area:
        ax.plot(contour[:, 1], contour[:, 0], linewidth=1.5, color='green')

ax.set_title("image_color")
plt.show()
```

```python
#Câu 10: Đếm số vật có trong ảnh


image_path = "anhtest.jpg"
image = io.imread(image_path, as_gray=True)

smoothed_image = filters.gaussian(image, sigma=0.1)

threshold_value = filters.threshold_otsu(smoothed_image)
binary_image = smoothed_image > threshold_value

binary_image = morphology.opening(binary_image, morphology.disk(1))

binary_image = morphology.closing(binary_image, morphology.disk(1))

labeled_image = measure.label(binary_image)

object_properties = measure.regionprops(labeled_image)

min_object_area = 10000  
max_object_area = 700000 
large_objects = [obj for obj in object_properties if min_object_area < obj.area < max_object_area]

num_large_objects = len(large_objects)

fig, ax = plt.subplots()
ax.imshow(image, cmap=plt.cm.gray)

for obj in large_objects:
    minr, minc, maxr, maxc = obj.bbox
    rect = plt.Rectangle((minc, minr), maxc - minc, maxr - minr,
                         fill=False, edgecolor='green', linewidth=2)
    ax.add_patch(rect)

ax.set_title(f"Detected Objects: {num_large_objects}")
plt.show()
```
