```python
#CAU_1
# Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load your own image
image_path = "anhtest.jpg"  # Replace with the actual path to your image
image = Image.open(image_path)

# Make some damaged pixels in the image
image_array = np.array(image)  # Convert image to NumPy array
image_array[100:300, 200:400] = [0, 0, 0]  # Example: Setting a rectangular region to black

# Create a mask for the damaged pixels
mask = np.zeros(image_array.shape[:2], dtype=bool)
mask[100:300, 200:400] = True  # Example: Defining a rectangular region as damaged

# Restore the original image using the mask
restored_image = np.copy(image_array)
restored_image[mask] = np.median(restored_image, axis=(0, 1))  # Example: Replacing damaged pixels with median RGB values

# Display the original and restored images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image_array)
axes[0].set_title('Original Image')
axes[0].axis('on')
axes[1].imshow(restored_image)
axes[1].set_title('Restored Image')
axes[1].axis('on')
plt.show()
```

```python
#CAU_2
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('on')
    
def plot_comparison(img_original, img_filtered, img_title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 8), sharex=True, sharey=True)
    ax1.imshow(img_original, cmap=plt.cm.gray)
    ax1.set_title('Original')
    ax1.axis('on')
    ax2.imshow(img_filtered, cmap=plt.cm.gray)
    ax2.set_title(img_title_filtered)
    ax2.axis('on')
    
import matplotlib.pyplot as plt
import numpy as np
from skimage.restoration import inpaint
from skimage.transform import resize
from skimage import color
image_with_logo = plt.imread('anhtestc3.jpg')

# Initialize the mask
mask = np.zeros(image_with_logo.shape[:-1])

# Set the pixels where the logo is to 1
mask[1050:1550, 250:1550] = 1
mask[2000:2200, 250:1000] = 1
mask[1700:2200, 900:1550] = 1
mask[1700:2200, 1582:] = 1
# Apply inpainting to remove the logo
image_logo_removed = inpaint.inpaint_biharmonic(image_with_logo,
                                                mask,
                                                channel_axis=-1)

# Show the original and logo removed images
plot_comparison(image_with_logo, image_logo_removed, 'Image with dots removed')
```

```python
#CAU_6
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def show_image(image, title='Image', cmap_type=None):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('on')
    plt.show()

def calculate_pixels(image_path):
    try:
        # Open the image using PIL
        image = Image.open(image_path)
        
        # Convert the image to a NumPy array
        image_array = np.array(image)
        
        # Get the shape of the image
        height, width, channels = image_array.shape
        
        # Calculate the number of pixels
        pixels = height * width
        
        return pixels
    
    except FileNotFoundError:
        print("Image file not found.")
    except Exception as e:
        print("An error occurred:", str(e))

# Provide the path to your image file
image_path = "anhtest.jpg"

# Calculate the number of pixels
num_pixels = calculate_pixels(image_path)

# Print the result
print("Number of pixels in the image:", num_pixels)

# Display the image
image = np.array(Image.open(image_path))
show_image(image)
```

```python
#CAU_8
import matplotlib.pyplot as plt
from skimage import io, filters, measure

# Load the image
image_path = "anhtest_binary.jpg"
image = io.imread(image_path, as_gray=True)

# Apply a threshold to convert the image to binary
threshold_value = filters.threshold_otsu(image)
binary_image = image > threshold_value

# Find contours in the binary image
contours = measure.find_contours(binary_image, 0.2)

# Filter contours based on size to detect large dots
min_dot_area = 100  # Minimum area of a dot (adjust as needed)

# Display the image and the detected dots
fig, ax = plt.subplots()
ax.imshow(image, cmap=plt.cm.gray)

for contour in contours:
    contour_area = len(contour)
    if contour_area > min_dot_area:
        ax.plot(contour[:, 1], contour[:, 0], linewidth=2, color='green')

ax.set_title("image_binary")
plt.show()
```

```python
#C_10
import matplotlib.pyplot as plt
from skimage import io, filters, measure, morphology

# Load the image
image_path = "anhtest.jpg"
image = io.imread(image_path, as_gray=True)

# Apply Gaussian smoothing to remove noise
smoothed_image = filters.gaussian(image, sigma=0.1)

# Apply a threshold to convert the image to binary
threshold_value = filters.threshold_otsu(smoothed_image)
binary_image = smoothed_image > threshold_value

# Remove small noise using morphological opening
binary_image = morphology.opening(binary_image, morphology.disk(1))

# Perform morphological closing to fill gaps and remove thin lines
binary_image = morphology.closing(binary_image, morphology.disk(1))

# Find connected components in the binary image
labeled_image = measure.label(binary_image)

# Calculate the region properties of the connected components
object_properties = measure.regionprops(labeled_image)

# Filter objects based on size to detect large objects
min_object_area = 10000  # Minimum area of an object (adjust as needed)
max_object_area = 700000 # Maximum area of an object (adjust as needed)
large_objects = [obj for obj in object_properties if min_object_area < obj.area < max_object_area]

# Count the number of large objects
num_large_objects = len(large_objects)

# Display the image and the detected objects
fig, ax = plt.subplots()
ax.imshow(image, cmap=plt.cm.gray)

for obj in large_objects:
    minr, minc, maxr, maxc = obj.bbox
    rect = plt.Rectangle((minc, minr), maxc - minc, maxr - minr,
                         fill=False, edgecolor='green', linewidth=2)
    ax.add_patch(rect)

ax.set_title(f"Detected Large Objects: {num_large_objects}")
plt.show()
```
